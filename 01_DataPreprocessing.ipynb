# carComment 处理

## 过滤掉该数据集

```python
import pandas as pd
```

```python
pd_car_comm = pd.read_csv('./rowData/carComment_pso_neg_03.csv')
pd_car_comm
```

```python
car_pos = pd_car_comm[pd_car_comm.label == 1]
car_neg = pd_car_comm[pd_car_comm.label == 0]
```

```python
car_pos_comments = car_pos['text']
car_neg_comments = car_neg['text']
```

```python
out = open('./data/pos_car.csv', 'w')
out.write("content" + '\n')
for comm in car_pos_comments:
    out.write(comm + '\n')

out = open('./data/neg_car.csv', 'w')
out.write("content" + '\n')
for comm in car_neg_comments:
    out.write(comm + '\n')
```

# goodsComment_02 处理

```python
pd_goods_comm_02 = pd.read_csv('./rowData/goodsComment_pos_neg_02.txt', sep='\t', names=['order', 'content', 'label'])
pd_goods_comm_02.head()
```

```python
# 这个数据集有三个标签 0：中性（模糊） 1:积极 2:消极，  这里只取 标签为 1、2
goods_02_pos = pd_goods_comm_02[pd_goods_comm_02.label == 1]
goods_02_neg = pd_goods_comm_02[pd_goods_comm_02.label == 2]
```

```python
goods_02_neg_comments = goods_02_neg['content']
goods_02_pos_comments = goods_02_pos['content']
```

```python
out = open('./data/pos_goods_02.csv', 'w')
for comm in goods_02_pos_comments:
    out.write(str(comm) + '\n')

out = open('./data/neg_goods_02.csv', 'w')
for comm in goods_02_neg_comments:
    out.write(str(comm) + '\n')
```

```python
pd_neg_goods_01 = pd.read_csv('./data/neg_goods_01.csv', sep='\t', names=['content'])
pd_neg_goods_02 = pd.read_csv('./data/neg_goods_02.csv', sep='\t', names=['content'])


pd_pos_goods_01 = pd.read_csv('./data/pos_goods_01.csv', sep='\t', names=['content'])
pd_pos_goods_02 = pd.read_csv('./data/pos_goods_02.csv', sep='\t', names=['content'])
```

```python
neg = []

for i in pd_neg_goods_01.content:
    neg.append(i)
for i in pd_neg_goods_02.content:
    neg.append(i)
    
pos = []

for i in pd_pos_goods_01.content:
    pos.append(i)
for i in pd_pos_goods_02.content:
    pos.append(i)
```

```python
out = open('./data/neg_goods.csv', 'w')
out.write('content' + '\n')
for comm in neg:
    out.write(str(comm) + '\n')
    
out = open('./data/pos_goods.csv', 'w')
out.write('content' + '\n')
for comm in pos:
    out.write(str(comm) + '\n')
```
